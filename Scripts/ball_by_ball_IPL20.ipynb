### match data parser method

```python
def match_data_parser(match_data):
    # balltype
    for i in match_data.bbcode.values:
        if "%SPEED%" in i:
            ball_type.append((i.split("%SPEED%", 1)[1]))
        else:
            ball_type.append("-")

    
    #bowler's info
    for i1 in match_data.bowlerId:
        for count1,player in enumerate(df_squads.players, 0):
            for j in player:
                if j['id'] == int(i1):
                    bowling_team.append(df_squads[df_squads.index == count1]['team.fullName'].unique()[0])
                    bowler.append(j['shortName'])
                    
                    if j['rightArmedBowl'] == False:
                        right_left_arm.append("Left")
                    else:
                        right_left_arm.append("Right")
                        
                    if ('bowlingStyle' in j):
                        bowling_style.append(j['bowlingStyle'])
                    else:
                        bowling_style.append("-")
                        
    #batsman's info
    for i2 in match_data.facingBatsmanId:
        
        for count,player in enumerate(df_squads.players,0):
            for j2 in player:
                if j2['id'] == int(i2):
                    batting_team.append(df_squads[df_squads.index == count]['team.fullName'].unique()[0])
                    batsman.append(j2['shortName'])
                    #print(j2['shortName'])
                    if j2['rightHandedBat'] == False:
                        batsman_batting_style.append("Left_handed")
                    else:
                        batsman_batting_style.append("Right_handed")
                
    #non-striker                
    for i3 in match_data.nonfacingBatsmanId:
        for player in df_squads.players:
            for j3 in player:
                if j3['id'] == int(i3):
                    non_striker.append(j3['shortName'])
                    
    
    # Wicket's data
    for i4 in match_data.bbcode:
        if "WICKET" in i4:
            
            wicket.append(i4.split("[b]WICKET!!! ", 1)[1].split("[/b]",1)[0].split(" is out", 1)[0])
            dismissal_comment.append(i4)
            #type_dismiss = i4.split("out[/b] ", 1)[1]
            #if "%SPEED%" in type_dismiss:
            #    dismissal = type_dismiss.split("%SPEED%", 1)[0]
            #    dismissal_comment.append(dismissal.split(",",1)[0])
            #    dismissal_kind.append(dismissal.split(" ",1)[0])
        
        else:
            wicket.append("-")
            dismissal_comment.append("-")
            #dismissal_kind.append("-")
            
    # Batsman_runs, Extra runs
    for i5 in match_data.thisOver:
        #print(i5)
        #print(match_data[match_data.thisOver == i5].bbcode)
        if i5 == '.':
            batsman_runs.append(0)
            extra_runs.append(0)
            
        elif i5 =='W': # wickets
            batsman_runs.append(0)
            extra_runs.append(0)
            
        elif i5 =='1W': # wickets
            batsman_runs.append(0)
            extra_runs.append(0)
            
        elif 'Wd' in i5: # legby
            
            if re.findall('\d*',i5)[0] == '':
                batsman_runs.append(0)
                extra_runs.append(1)
            if re.findall('\d*',i5)[0] != '':
                wd = re.findall('\d*',i5)[0]
                
                batsman_runs.append(0)
                extra_runs.append(int(wd))
            

            
        elif 'Lb' in i5: # legbye
            
            if re.findall('\d*',i5)[0] == '':
                batsman_runs.append(0)
                extra_runs.append(1)
            if re.findall('\d*',i5)[0] != '':
                lb = re.findall('\d*',i5)[0]
                
                batsman_runs.append(0)
                extra_runs.append(int(lb))
                
                
        elif 'B' in i5: # bye runs
            
            if re.findall('\d*',i5)[0] == '':
                batsman_runs.append(0)
                extra_runs.append(1)
            if re.findall('\d*',i5)[0] != '':
                bye = re.findall('\d*',i5)[0]
                print(int(bye))
                batsman_runs.append(0)
                extra_runs.append(int(bye))
            
        
            
        elif 'Nb' in i5: # Noball runs
            
            if re.findall('\d*',i5)[0] == '':
                batsman_runs.append(0)
                extra_runs.append(1)
            if re.findall('\d*',i5)[0] != '':
                nb = re.findall('\d*',i5)[0]
                
                batsman_runs.append(0)
                extra_runs.append(int(nb))
            
            #print(match_data[match_data.thisOver == i5].bbcode)
        else:
            batsman_runs.append(int(i5))
            extra_runs.append(0)

    match_data["batsman_runs"] = batsman_runs
    match_data["extra_runs"] = extra_runs
    
    # Inning , over and ball
    for i6, i7, i8 in zip(match_data["progress.innings"], match_data["progress.over"], match_data["progress.ball"]):
        inning.append(int(i6))
        over.append(int(i7))
        ball.append(int(i8))

    

                    
    # Data fillup
    match_data["ball_type"] = ball_type
    if 'speed' in match_data.columns.values:
        match_data["ball_speed"] = round((match_data.speed *3.6),1)
    else:
        match_data["ball_speed"] = "-"
        
    match_data["inning"] = inning
    match_data["over"] = over
    match_data["ball"] = ball
    match_data["bowler"] = bowler
    match_data["right_left_arm"] = right_left_arm
    match_data["bowling_style"] = bowling_style
    match_data["batsman"] = batsman
    match_data["batsman_batting_style"] = batsman_batting_style
    match_data["non_striker"] = non_striker
    match_data["batting_team"] =batting_team
    match_data["bowling_team"] = bowling_team
    match_data["player_dismissed"] = wicket
    match_data["dismissal_comment"] = dismissal_comment
    #match_data["dismissal_kind"] = dismissal_kind
    IPL_2020 = match_data[['inning', 'batting_team', 'bowling_team','over', 'ball', 'batsman', 
            'non_striker', 'bowler', 'ball_type', 'ball_speed', 'right_left_arm',
            'bowling_style', 'batsman_batting_style',  'thisOver','batsman_runs','extra_runs',
            'player_dismissed',  'bbcode']]#, 'dismissal_comment', 'dismissal_kind', 
    return IPL_2020
    
```
